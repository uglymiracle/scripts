```python
#магический шар предсказаний
from tkinter import *
window = Tk()
window['bg'] = 'SkyBlue2'
window.title("Магический шар")
window.geometry('700x100')
window.wm_attributes('-alpha', 0.7)
answers = ['Бесспорно', "Мне кажется - да", 'Пока неясно, попробуй снова', "Даже не думай", 'Предрешено', 'Вероятнее всего', 'Спроси позже', 'Мой ответ - нет', 'Никаких сомнений', 'Хорошие перспективы', 'Лучше не рассказывать', 'По моим данным - нет', 'Определённо да', 'Знаки говорят - да', 'Сейчас нельзя предсказать', 'Перспективы не очень хорошие', 'Можешь быть уверен в этом', 'Да', 'Сконцентрируйся и спроси опять', 'Весьма сомнительно']
questions = ['Спрашивай', 'Задавай свой вопрос', 'Что ты хочешь узнать?', 'Что же волнует твою душу?', 'В чем же тебе нужен мой совет?', 'Чем я могу тебе помочь?']
import random  
    
def sendget():
    global names
    names = name.get()
    lbl.destroy()
    lbl2.destroy()
    btn_name.destroy()
    name.destroy()
    global lbl_names
    lbl_names = Label(window, text='Приятно познакомиться, ' + names, font=("Helvetica", 20), bg="SkyBlue2", fg="black")
    lbl_names.grid(column=0, row=0)
    question()

def question():
    global lbl_question, quest, btn_quest
    lbl_question = Label(window, text=random.choice(questions), font=("Helvetica", 20), bg="SkyBlue2", fg="black")
    lbl_question.grid(column=0, row=1, columnspan=4)
    quest = Entry(window, width=30)
    quest.focus()
    quest.grid(column=0, row=3)
    btn_quest = Button(window, text="Отправить", font=("Helvetica", 20), bg="SkyBlue2", fg="black", command=sendget2)
    btn_quest.grid(column=1, row=3)

def sendget2():
    quests = quest.get()
    lbl_names.destroy()
    lbl_question.destroy()
    btn_quest.destroy()
    quest.destroy()
    global lbl_answer
    lbl_answer = Label(window, text=random.choice(answers), font=("Helvetica", 20), bg="SkyBlue2", fg="black")
    lbl_answer.grid(column=0, row=0)
    cont()
    
def cont():
    global lbl_cont, btn_no, btn_yes
    lbl_cont = Label(window, text = names + ',' + ' желаешь спросить что-то еще? (да, нет)',  font=("Helvetica", 20), bg="SkyBlue2", fg="black")
    lbl_cont.grid(column=0, row=1)
    btn_yes = Button(window, text="Да", font=("Helvetica", 20), bg="SkyBlue2", fg="black", command=clicked_yes)
    btn_yes.grid(column=0, row=2)
    btn_no = Button(window, text="Нет", font=("Helvetica", 20), bg="SkyBlue2", fg="black", command=clicked_no)
    btn_no.grid(column=1, row=2) 
    
def clicked_yes():
    lbl_answer.destroy()
    lbl_cont.destroy()
    btn_no.destroy() 
    btn_yes.destroy()
    question()

def clicked_no():
    lbl_answer.destroy()
    lbl_cont.destroy()
    btn_no.destroy() 
    btn_yes.destroy()
    lbl_no = Label(window, text='Слушай своё сердце. Еще увидимся...', font=("Helvetica", 20), bg="SkyBlue2", fg="black")
    lbl_no.grid()
    
lbl = Label(window, text="Привет, мир, я магический шар, и я знаю ответ на любой твой вопрос.", font=("Helvetica", 20), bg="SkyBlue2", fg="black")
lbl.grid(column=0, row=0, columnspan=8) 
lbl2 = Label(window, text='Назови свое имя', font=("Helvetica", 20), bg="SkyBlue2", fg="black")
lbl2.grid(column=0, row=1) 

name = Entry(window, width=30)
name.focus()
name.grid(column=0, row=6)

btn_name = Button(window, text="Отправить", font=("Helvetica", 20), bg="SkyBlue2", fg="black", command=sendget)
btn_name.grid(column=1, row=6)

window.mainloop()
```
